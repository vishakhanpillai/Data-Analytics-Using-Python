```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.xls')
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df = df.drop(columns=["Cabin"])
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Counts (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival by Gender (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.title("Survival by Passenger Class (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Embarked", data=df)
plt.title("Survival by Embarked (S vs others)")
plt.show()
```
