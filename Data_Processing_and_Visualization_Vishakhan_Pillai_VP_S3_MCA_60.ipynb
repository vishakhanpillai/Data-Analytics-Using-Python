Titanic Dataset processing and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.xls')
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df = df.drop(columns=["Cabin"])
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Counts (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival by Gender (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Pclass", data=df)
plt.title("Survival by Passenger Class (0 = Died, 1 = Survived)")
plt.show()
```

```python
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Embarked", data=df)
plt.title("Survival by Embarked (S vs others)")
plt.show()
```

Air Quality Dataset Preprocessing and Visualization

```python
df2 = pd.read_excel('AirQualityUCI.xlsx')
df2
```

```python
print(df2.shape)
print(df2.info())
print(df2.head())
```

```python
df2.isnull().sum()
```

```python
df2.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df2['Date'], df2['CO(GT)'], label="CO")
plt.plot(df2['Date'], df2['NO2(GT)'], label="NO2")
plt.plot(df2['Date'], df2['PT08.S5(O3)'], label="O3 (Sensor)")
plt.xlabel("Date")
plt.ylabel("Concentration")
plt.title("Air Pollutants Over Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df2.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Pollutants")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df2[['CO(GT)','NO2(GT)','PT08.S5(O3)']])
plt.title("Boxplot of Pollutant Concentrations")
plt.show()
```

```python
plt.figure(figsize=(14,7))

plt.plot(df2['Date'], df2['CO(GT)'], label="CO", alpha=0.7)
plt.plot(df2['Date'], df2['NO2(GT)'], label="NO2", alpha=0.7)
plt.plot(df2['Date'], df2['PT08.S5(O3)'], label="O3 (Sensor)", alpha=0.7)

plt.plot(df2['Date'], df2['T'], label="Temperature (Â°C)", linestyle="--")
plt.plot(df2['Date'], df2['RH'], label="Relative Humidity (%)", linestyle="--")
plt.plot(df2['Date'], df2['AH'], label="Absolute Humidity", linestyle="--")

plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Pollutants & Weather Conditions Over Time")
plt.legend()
plt.show()
```
